```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/kaggle/input/smokers-health-data/smoking_health_data_final.csv")
```

```python
# Distribution of Age
plt.hist(data['age'], bins=20, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Age vs. Heart Rate
plt.scatter(data['age'], data['heart_rate'], alpha=0.5, color='orange')
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.title('Age vs. Heart Rate')
plt.show()
```

```python
# Boxplot of Cigarettes Per Day
sns.boxplot(data['cigs_per_day'])
plt.xlabel('Cigarettes Per Day')
plt.title('Distribution of Cigarettes Per Day')
plt.show()
```

```python
# Scatterplot of Cholesterol vs. Age
plt.scatter(data['age'], data['chol'], alpha=0.5, color='purple')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age vs. Cholesterol')
plt.show()
```

```python
# Violin Plot of Cholesterol by Sex:
sns.violinplot(x=data['sex'], y=data['chol'])
plt.xlabel('Sex')
plt.ylabel('Cholesterol')
plt.title('Cholesterol Levels by Sex')
plt.show()
```

```python
# Average Heart Rate by Smoking Status
avg_heart_rate = data.groupby('current_smoker')['heart_rate'].mean()
avg_heart_rate.plot(kind='bar', color=['blue', 'red'])
plt.xlabel('Smoking Status')
plt.ylabel('Average Heart Rate')
plt.title('Average Heart Rate by Smoking Status')
plt.xticks([0, 1], ['Non-Smoker', 'Smoker'], rotation=0)
plt.show()
```

```python
# Average Cholesterol by Smoking Status
avg_chol = data.groupby('current_smoker')['chol'].mean()
avg_chol.plot(kind='bar', color=['blue', 'red'])
plt.xlabel('Smoking Status')
plt.ylabel('Average Cholesterol')
plt.title('Average Cholesterol by Smoking Status')
plt.xticks([0, 1], ['Non-Smoker', 'Smoker'], rotation=0)
plt.show()
```

```python
# Heart Rate by Gender
sns.boxplot(x=data['sex'], y=data['heart_rate'])
plt.xlabel('Gender')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Distribution by Gender')
plt.show()
```

```python
# Cigarettes Per Day vs. Cholesterol
plt.scatter(data['cigs_per_day'], data['chol'], alpha=0.5, color='red')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Cholesterol')
plt.title('Cigarettes Per Day vs. Cholesterol')
plt.show()
```

```python
# Age Group Distribution
age_groups = pd.cut(data['age'], bins=range(20, 81, 10))
age_groups.value_counts().sort_index().plot(kind='bar', color='green')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Age Groups')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cholesterol Levels Distribution
plt.hist(data['chol'], color='purple')
plt.xlabel('Cholesterol Levels')
plt.ylabel('Frequency')
plt.title('Distribution of Cholesterol Levels')
plt.show()
```
